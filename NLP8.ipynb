```python
# Batch 1: Import libraries and define sample text data

import math
from collections import Counter
from itertools import islice

# Sample documents (same style as in your slides)
documents = [
    "I like this product. Do you like this product?",
    "This product is not good",
    "The product is good",
    "A bad product",
    "Average product"
]

print("Documents Loaded Successfully!")
```

```python
# Batch 2: Basic preprocessing (lowercase + remove punctuation)

import re

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    return text.split()

tokenized_docs = [preprocess(doc) for doc in documents]

print("Tokenized Documents:")
for doc in tokenized_docs:
    print(doc)
```

```python
# Batch 3: Calculate Term Frequency

def compute_tf(doc_tokens):
    tf_dict = {}
    total_words = len(doc_tokens)
    counts = Counter(doc_tokens)

    for word, count in counts.items():
        tf_dict[word] = count / total_words

    return tf_dict

tf_list = [compute_tf(doc) for doc in tokenized_docs]

print("Term Frequency for each document:")
for tf in tf_list:
    print(tf)
```

```python
# Batch 4: Calculate IDF

def compute_idf(tokenized_docs):
    N = len(tokenized_docs)
    idf_dict = {}

    all_words = set(word for doc in tokenized_docs for word in doc)

    for word in all_words:
        containing_docs = sum(1 for doc in tokenized_docs if word in doc)
        idf_dict[word] = math.log(N / containing_docs)

    return idf_dict

idf = compute_idf(tokenized_docs)

print("Inverse Document Frequency:")
print(idf)
```

```python
# Batch 5: Compute TF-IDF

def compute_tfidf(tf_list, idf):
    tfidf_list = []

    for tf in tf_list:
        doc_tfidf = {}
        for word, value in tf.items():
            doc_tfidf[word] = value * idf[word]
        tfidf_list.append(doc_tfidf)

    return tfidf_list

tfidf = compute_tfidf(tf_list, idf)

print("TF-IDF for each document:")
for doc in tfidf:
    print(doc)
```

```python
# Batch 6: Generate N-grams

def generate_ngrams(tokens, n):
    return list(zip(*(islice(tokens, i, None) for i in range(n))))

# Example using first document
sample_tokens = tokenized_docs[0]

print("Unigrams:")
print(sample_tokens)

print("\nBigrams:")
print(generate_ngrams(sample_tokens, 2))

print("\nTrigrams:")
print(generate_ngrams(sample_tokens, 3))
```

```python
# Batch 7: Frequency of N-grams

def ngram_frequency(tokens, n):
    ngrams = generate_ngrams(tokens, n)
    return Counter(ngrams)

print("Bigram Frequency (Doc 1):")
print(ngram_frequency(sample_tokens, 2))

print("\nTrigram Frequency (Doc 1):")
print(ngram_frequency(sample_tokens, 3))
```
